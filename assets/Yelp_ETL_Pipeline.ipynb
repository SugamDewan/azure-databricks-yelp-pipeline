```python
# 1. Define the storage details
storage_account_name = "yelpsg"
storage_account_key = "[HIDDEN FOR SECURITY]="
storage_container_name = "sourcecontainer"

# 2. Securely provide the key to Spark
spark.conf.set(
    f"fs.azure.account.key.{storage_account_name}.dfs.core.windows.net",
    storage_account_key
)

# 3. Test the connection by listing files in your container
path = f"abfss://{storage_container_name}@{storage_account_name}.dfs.core.windows.net/"
display(dbutils.fs.ls(path))
```

```python
# 1 Setting the pathn to the specific file
file_name = "yelp_academic_dataset_business.json"
file_path = f"{path}{file_name}"

# 2 Read the Data
df = spark.read.format("json").load(file_path)

# 3 Show the data schema (columns)
print("Data Schema:")
df.printSchema()

print("Preview of Data:")
display(df.limit(10))
```

```python
# 1 Count the total rows
total_count = df.count()
print(f"Total records: {total_count}")

# 2 Looking at the data structure (Schema) again
df.printSchema()

# 3 Check for duplicates
duplicate_count = df.count() - df.dropDuplicates().count()
print(f"Number of duplicate rows: {duplicate_count}")
```

```python
from pyspark.sql.functions import col

# 1 Create a new "Silver DataFrame with transformations
silver_df = df.filter(col("is_open") == 1)   \
    .select(
        col("name"),
        col("city"),
        col("state"),
        col("stars").alias("rating"),
        col("review_count")
    )
# 2 Show the clean data
print("Here is the Cleaned (Silver) Data:")
display(silver_df.limit(10))
```

```python
# We are saving the clean data in the storage account as a delta table
# We are creating a new folder in the storage account called "silver_yelp_businesses"
save_path = f"{path}silver_yelp_businesses"
silver_df.write.format("delta").mode("overwrite").save(save_path)

print(f"Success! Clean data saved to: {save_path}")
```

```python
from pyspark.sql.functions import desc, count
# 1 Creating the Gold DataFrame (Transformation)
gold_df = silver_df.groupBy("city") \
    .agg(count("name").alias("business_count")) \
    .orderBy(col("business_count").desc())
    
# 2 Show the top 10 cities 
print("Gold Layer: Top Cities by Business Count")
display(gold_df.limit(10))

# 3 Saving this Gold table (Load)
gold_path = f"{path}gold_city_stats"
gold_df.write.format("delta").mode("overwrite").save(gold_path)

print(f"Success! Gold analytics saved to: {gold_path}")
```

```python
#1 Load the data from your storageback into Spark
silver_df = spark.read.format("delta").load(f"{path}silver_yelp_businesses")
gold_df = spark.read.format("delta").load(f"{path}gold_city_stats")

#2 Register them as "Temporary Views"
#this will let us run SQL queries 
silver_df.createOrReplaceTempView("silver_yelp_businesses")
gold_df.createOrReplaceTempView("gold_city_counts")

print("Success! SQL Queries are ready to run")
```
